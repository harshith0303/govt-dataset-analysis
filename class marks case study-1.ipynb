```python
#loading dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("class_marks.csv")
```

```python
data
```

```python
# Filling Missing Values with 0
data.fillna(0,inplace=True)
data
```

```python
#checking missing vaalues in dataset
print("Missing Values Before Cleaning:")
print(data.isnull().sum())
```

```python
data
```

```python
# Displaying the First Few Rows of the Cleaned Dataset
print("Cleaned Dataset:")
print(data.head())
```

```python
#highest and lowest scores
highest_score = data['Total'].max()
highest_scorer = data[data['Total'] == highest_score]
lowest_score = data['Total'].min()
lowest_scorer = data[data['Total'] == lowest_score]
print("Highest Scorer(s):")
print(highest_scorer)
print("Lowest Scorer(s):")
print(lowest_scorer)
```

```python
data.hist()
```

```python
# Calculating average marks for each column
average_marks = data.mean(numeric_only=True)
print("Average Marks for Each Question and Total:")
print(average_marks)
```

```python
#avg marks.for each question
average_marks.drop('Total').plot(kind='bar', color='teal', figsize=(10, 6))
plt.title('Average Marks for Each Question')
plt.xlabel('Question')
plt.ylabel('Average Marks')
plt.xticks(rotation=45)
plt.show()
average_total = average_marks['Total']
print(f"Average Total Score: {average_total:.2f}")
```

```python
#combining questions
data['Q1']=data["Q1aM4"]+data["Q1bM6"]
data['Q2']=data['Q2aM6']+data['Q2bM4']
data['Q3']=data['Q3aM5']+data['Q3bM5']
data['Q4']=data['Q4aM3']+data['Q4bM7']
data['Q6']=data['Q6aM4']+data['Q6bM6']
data
```

```python
#distribution of total scores
sns.histplot(data['Total'], kde=True, bins=10, color='blue')
plt.title('Distribution of Total Scores')
plt.xlabel('Total Score')
plt.ylabel('Frequency')
plt.show()
```

```python
#hghest and lowest scores
highest_score = data['Total'].max()
lowest_score = data['Total'].min()
score_data = pd.DataFrame({
    'Type': ['Highest Score', 'Lowest Score'],
    'Score': [highest_score, lowest_score]
})
sns.barplot(x='Type', y='Score', data=score_data, palette='Set2')
plt.title('Highest and Lowest Total Scores')
plt.ylabel('Score')
plt.show()
```

```python
# Reshaping data for easier visualization
melted_data = data.melt(id_vars=['Total'], var_name='Question', value_name='Score')
melted_data['Score'] = pd.to_numeric(melted_data['Score'], errors='coerce')
plt.figure(figsize=(12, 6))
sns.boxplot(x='Question', y='Score', data=melted_data, palette='Set2')
plt.title('Boxplot of Scores by Question')
plt.xlabel('Question')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.show()
```

```python
#categorising students into groups based on performance
bins = [0, 20, 40, 60]
labels = ['Low', 'Medium', 'High']
data['Performance'] = pd.cut(data['Total'], bins=bins, labels=labels)
sns.countplot(x='Performance', data=data, palette='muted')
plt.title('Performance Groups of Students')
plt.xlabel('Performance Group')
plt.ylabel('Count')
plt.show()
```

```python
#heatmap for questions and overall score
corr = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Questions and Total Score')
plt.show()
```

```python
# Setting Seaborn theme to white
sns.set_theme(style="white")
melted_data = data.melt(id_vars=['Total'], var_name='Question', value_name='Score')
melted_data['Score'] = pd.to_numeric(melted_data['Score'], errors='coerce')
plt.figure(figsize=(12, 6))
sns.boxplot(x='Question', y='Score', data=melted_data, palette='Set2')
plt.title('Boxplot of Scores by Question')
plt.xlabel('Question')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# Setting Seaborn theme to white
sns.set_theme(style="white")
sns.relplot(x='Q1aM4', y='Total', data=data, kind='scatter', height=6, aspect=1.5, hue='Total', palette='coolwarm')
plt.title('Scatter Plot: Q1aM4 vs Total Score')
plt.xlabel('Q1aM4 Marks')
plt.ylabel('Total Score')
plt.show()
```

```python
# Defining performance groups based on Total 
bins = [0, 20, 40, 60]
labels = ['Low', 'Medium', 'High']
data['Performance'] = pd.cut(data['Total'], bins=bins, labels=labels)
performance_counts = data['Performance'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(performance_counts, labels=performance_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set2"))
plt.title('Performance Distribution of Students')
plt.axis('equal')  
plt.show()
```
